```python
heroes = sorted(df['Enemy_pos1'].unique())
```

```python
df.sort_values(by='Enemy_pos1')
```

```python
winrate = {}

for hero in heroes:
    matches = df[df['Enemy_pos1'] == hero]
    matches_won = matches[matches["Result"] == 'Won Match']
    winrate = round(len(matches_won) / len(matches) * 100,2)
    
    print(str(winrate) + "% - " + hero)
    
```
