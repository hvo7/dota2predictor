```python
import pandas as pd
import numpy as np
```

```python
# Import data
df = pd.read_csv(r"C:\Users\henry\OneDrive - Louisiana State University\Desktop\Dota2Predictor\dota.csv")
```

```python
df.dropna(inplace=True)
```

```python
df.tail()
```

```python
enemy_heroes_pos1 = sorted(df['Enemy_pos1'].unique())
enemy_heroes_pos2 = sorted(df['Enemy_pos2'].unique())
enemy_heroes_pos3 = sorted(df['Enemy_pos3'].unique())
enemy_heroes_pos4 = sorted(df['Enemy_pos4'].unique())
enemy_heroes_pos5 = sorted(df['Enemy_pos5'].unique())
```

```python
# Calculate winrates against enemy position 1's
winrates_enemy_pos1 = {}

for hero in enemy_heroes_pos1:
    matches = df[df['Enemy_pos1'] == hero]
    matches_won = matches[matches["Result"] == 'Won Match']
    winrate = round(len(matches_won) / len(matches) * 100,2)

    # Add winrate and hero to dictionary
    winrates_enemy_pos1[hero] = winrate
```
